#**K-Nearest Neighbors**

Examples of K-Nearest Neighbors

K-Nearest Neighbors (KNN) is a simple and versatile algorithm used for both classification and regression tasks. It works by finding the 'k' training examples in the dataset that are closest to a given input data point and then making predictions based on the majority class (for classification) or the average of the 'k' nearest neighbors' values (for regression). Here are some examples of how K-Nearest Neighbors can be applied:

1. **Image Recognition:**
   - **Scenario:** Classifying images into different categories.
   - **Example:** Given a dataset of labeled images, KNN can be used to classify a new image by comparing it to the 'k' nearest neighbors in the training set.

2. **Credit Scoring:**
   - **Scenario:** Assessing the creditworthiness of individuals.
   - **Example:** Using features such as income, credit score, and debt, KNN can be applied to identify individuals with similar financial profiles and predict their creditworthiness.

3. **Recommendation Systems:**
   - **Scenario:** Recommending products or content based on user preferences.
   - **Example:** By comparing a user's preferences and behavior to those of similar users (neighbors), KNN can recommend products or content that the user's neighbors have liked.

4. **Healthcare:**
   - **Scenario:** Diagnosing medical conditions based on patient data.
   - **Example:** Using patient information like symptoms, medical history, and test results, KNN can help identify similar cases and assist in diagnosing medical conditions.

5. **Anomaly Detection:**
   - **Scenario:** Identifying outliers or anomalies in a dataset.
   - **Example:** KNN can be used to identify data points that deviate significantly from the majority of the dataset, making it useful for anomaly detection in areas such as fraud detection or network security.

6. **Geospatial Analysis:**
   - **Scenario:** Predicting property prices based on location and features.
   - **Example:** KNN can be applied to predict property prices by considering the prices of neighboring properties with similar characteristics and locations.

7. **Customer Segmentation:**
   - **Scenario:** Grouping customers based on their purchasing behavior.
   - **Example:** KNN can be used to cluster customers with similar buying patterns, helping businesses tailor marketing strategies for different customer segments.

8. **Environmental Monitoring:**
   - **Scenario:** Classifying plant or animal species based on environmental features.
   - **Example:** KNN can be applied to classify plant or animal species by comparing environmental features like temperature, humidity, and soil composition to those of neighboring regions.

These examples illustrate the versatility of K-Nearest Neighbors across various domains for both classification and regression tasks. The choice of 'k' and the distance metric used are crucial parameters that can be adjusted based on the specific problem and dataset characteristics.

#**Importing Data Set**

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

**Create KNN (K Neighrest Neighbour Classifier)**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
knn.predict([[4.8, 3.0, 1.5, 3.0]])
```
